```python
%sql
CREATE OR REPLACE TABLE my_catalog.my_schema.gold_customer_summary AS
SELECT 
    customer_id,
    COUNT(order_id) AS total_orders,
    SUM(amount) AS total_spent,
    AVG(amount) AS avg_spent
FROM my_catalog.my_schema.gold_sales
GROUP BY customer_id;
```

```python
%sql
select * from my_catalog.my_schema.gold_customer_summary
```
