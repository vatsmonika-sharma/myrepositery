```python
%sql


-- Add derived columns
CREATE OR REPLACE TABLE my_catalog.my_schema.gold_sales AS
SELECT 
    order_id,
    customer_id,
    product_id,
    amount,
    order_date,
    YEAR(order_date) AS year,
    MONTH(order_date) AS month,
    CASE 
        WHEN amount > 150 THEN 'High'
        ELSE 'Low'
    END AS sales_category
FROM my_catalog.my_schema.silver_sales;
```
