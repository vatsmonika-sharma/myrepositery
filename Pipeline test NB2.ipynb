```python
%sql
CREATE OR REPLACE TABLE my_catalog.my_schema.silver_sales AS
SELECT order_id,
       customer_id,
       product_id,
       amount,
       TO_DATE(order_date) AS order_date
FROM (
    SELECT *,
           ROW_NUMBER() OVER (PARTITION BY order_id ORDER BY order_date DESC) AS rn
    FROM my_catalog.my_schema.bronze_sales
) t
WHERE rn = 1;
```
